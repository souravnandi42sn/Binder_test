```python
#preprocess the customer_attributes
import re
ca=0
for i in t.index:
    if(not pd.isnull(t.loc[i,'customer_attributes'])):
        s=t.loc[i,'customer_attributes']
        s=s.replace("\n"," ")
        t.loc[i,'customer_attributes']=re.sub(r"\s+", " ", s)
        ca=ca+1
    else:
        t.loc[i,'customer_attributes']=" "
```
